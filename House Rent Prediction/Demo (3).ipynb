```python
data['Society'] = data['Society'].apply(lambda x: 'Other' if x in society_count_nan else x)
data['Society'] = data['Society'].apply(lambda x: x.upper())
```

```python
data['Location'] = data['Location'].apply(lambda x: str(x).strip())
data['Location'] = data['Location'].apply(lambda x: x.upper())
```

```python
power = []
for li in data['Amenities']:
    if 'Power Back Up' in li:
        power.append(1)
    else:
        power.append(0)
```

```python
data['Power Back Up'] = power
```

```python
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace("'Power Back Up',",""))
```

```python
lift = []
for li in data['Amenities']:
    if 'Lift' in li:
        lift.append(1)
    else:
        lift.append(0)
data['Lift'] = lift
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace("'Lift'",""))
```

```python
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace(","," "))
```

```python
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace("]",""))
```

```python
data['Amenities'] = data['Amenities'].apply(lambda x: x.strip())
data['Amenities']
```

```python
Swim = []
for li in data['Amenities']:
    if 'Swimming Pool' in li:
        Swim.append(1)
    else:
        Swim.append(0)
data['Swimming Pool'] = Swim
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace("'Swimming Pool'",""))
data['Amenities'] = data['Amenities'].apply(lambda x: x.strip())
```

```python
gym = []
for li in data['Amenities']:
    if 'Gymnasium' in li:
        gym.append(1)
    else:
        gym.append(0)
data['Gym'] = gym
data['Amenities'] = data['Amenities'].apply(lambda x: x.replace("'Gymnasium'",""))
data['Amenities'] = data['Amenities'].apply(lambda x: x.strip())
```

```python
data['ImpInfo'].value_counts()
```

```python
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("[",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("]",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("1Bath",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("'Semi-Furnished'",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("'Unfurnished'",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("'Furnished'",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace("'",""))
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.strip())
```

```python
data['ImpInfo'] = data['ImpInfo'].apply(lambda x: x.replace(",",""))
```

```python
x = data.drop(columns =['Rent'])
y = data['Rent']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse = False,handle_unknown='ignore'), ['Loc']), remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

```python
ridge = Ridge()
pipe = make_pipeline(column_trans, scaler,ridge)
pipe.fit(x_train,y_train)
y_pred_ridge = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

```python
lasso = Lasso()
pipe = make_pipeline(column_trans, scaler, lasso)
pipe.fit(x_train,y_train)
y_pred_ridge = pipe.predict(x_test)
```
